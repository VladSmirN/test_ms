```python
%%writefile yolo_facade.yaml
train: /kaggle/input/facade-detectio/facade_detection/train/images
val: /kaggle/input/facade-detectio/facade_detection/valid/images
names:  
  0: facade
```

```python
model = YOLO('yolov8n.pt')
#https://github.com/ultralytics/ultralytics/blob/main/ultralytics/yolo/cfg/default.yaml
results = model.train(data="/kaggle/working/yolo_facade.yaml", 
                      epochs=200, 
                      imgsz=256,
                      optimizer="Adam",
                      lr0=0.001,
                      batch=-1,
                      patience=15
                     )
```
